## **Data Preprocessing Functions**

```python
def cleaning_text(text):
  text = text.lower()
  text_clean = re.sub(r'\d+','',text)
  text_clean =re.sub(r'\s+',' ',text_clean)
  text_clean = text_clean.strip()
  return text_clean


def removehtml(text):
  try:
    return BeautifulSoup(text,"html.parser").get_text()
  except Exception as e:
    print(f"Error parsing HTML: {e}")
    return text

def filter_by_length(text,minl=15,maxl= 120):
  length = len(text.split())
  if(length>= minl and length <= maxl):
    return text
  else:
    return None

def lang_detect(text):
  try:
    return detect(text)
  except:
    return None


def spell_checker(tokens):
  spell = SpellChecker()
  return [spell.correction(word)if word not in spell else word for word in tokens]
```

```python
from datasets import Dataset
dataset = Dataset.from_pandas(df)

# Load the tokenizer
tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-es")

# Preprocess function to tokenize input and target text
def preprocess_function(examples):
    inputs = [ex for ex in examples["English"]]
    targets = [ex for ex in examples["Spanish"]]
    model_inputs = tokenizer(inputs, max_length=MAX_SEQUENCE_LENGTH, truncation=True, padding="max_length")

    # Tokenize targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=MAX_SEQUENCE_LENGTH, truncation=True, padding="max_length")

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Tokenize dataset
tokenized_datasets = dataset.map(preprocess_function, batched=True)
```

```python
# Load pre-trained MarianMT model
model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-es")

# Split dataset into training and validation
train_size = 0.8
train_testsplit = tokenized_datasets.train_test_split(test_size=(1 - train_size))
train_dataset = train_testsplit["train"]
eval_dataset = train_testsplit["test"]

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    save_total_limit=3,
    load_best_model_at_end=True,  # Load the best model when finished
    metric_for_best_model="eval_loss",  # Metric to monitor for early stopping
    greater_is_better=False,  # Lower loss is better
      # Evaluate every epoch
    save_strategy="epoch"
)

# Define Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,callbacks=[EarlyStoppingCallback(early_stopping_patience=2)],
)
trainer.train()
```

```python
import streamlit as st
from transformers import MarianMTModel, MarianTokenizer
import torch
import pdfplumber
import time

# Load Pretrained Models and Tokenizers
@st.cache_resource
def load_models():
    model_en_to_es = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-es")
    tokenizer_en_to_es = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-es")

    model_es_to_en = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-es-en")
    tokenizer_es_to_en = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-es-en")

    return (model_en_to_es, tokenizer_en_to_es), (model_es_to_en, tokenizer_es_to_en)

# Load the models
(en_to_es_model, en_to_es_tokenizer), (es_to_en_model, es_to_en_tokenizer) = load_models()

# Translation Function
def translate(input_text, model, tokenizer):
    inputs = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        translated_tokens = model.generate(inputs["input_ids"], max_length=128, num_beams=4, early_stopping=True)
    return tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

# Extract text from PDF
def extract_text_from_pdf(pdf_file):
    with pdfplumber.open(pdf_file) as pdf:
        text = ""
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Streamlit Interface
st.title("Bidirectional Language Translation App 🌐")
st.write("Translate between **English ↔ Spanish** and upload PDF files for bulk translation.")

# Tabs for Text and PDF Translation
tab1, tab2 = st.tabs(["🔤 Text Translation", "📄 PDF Translation"])

with tab1:
    st.subheader("🔤 Text Translation")
    option = st.radio("Select Translation Direction:", ["English to Spanish", "Spanish to English"])

    input_text = st.text_area("Enter text for translation:")
    if st.button("Translate"):
        if not input_text.strip():
            st.warning("Please enter some text to translate.")
        else:
            start_time = time.time()
            if option == "English to Spanish":
                result = translate(input_text, en_to_es_model, en_to_es_tokenizer)
            else:
                result = translate(input_text, es_to_en_model, es_to_en_tokenizer)
            end_time = time.time()

            st.success("Translation Completed!")
            st.write(f"**Translated Text:**{result}")
            st.write(f"⏱️ **Time Taken:** {end_time - start_time:.2f} seconds")

with tab2:
    st.subheader("📄 PDF Translation")
    option = st.radio("Select Translation Direction for PDF:", ["English to Spanish", "Spanish to English"], key="pdf_direction")

    uploaded_pdf = st.file_uploader("Upload your PDF file:", type=["pdf"])
    if uploaded_pdf:
        st.info("Extracting text from PDF...")
        extracted_text = extract_text_from_pdf(uploaded_pdf)
        st.text_area("Extracted PDF Content:", extracted_text, height=200)

        if st.button("Translate PDF"):
            if not extracted_text.strip():
                st.warning("No text found in the PDF file. Please upload a valid file.")
            else:
                start_time = time.time()
                # Split text into smaller chunks for translation
                chunks = [content[i:i + 100] for i in range(0, len(content), 100)]  # Adjust chunk size as needed

                # Initialize translated text
                translated_text = ""
                # Translate each chunk
                for chunk in chunks:
                  if option == "English to Spanish":
                    translated_text += translate(chunk, en_to_es_model, en_to_es_tokenizer)
                  else:
                    translated_text += translate(chunk, es_to_en_model, es_to_en_tokenizer)
                end_time = time.time()

                st.success("PDF Translation Completed!")
                st.text_area("Translated PDF Content:", translated_text, height=300)
                st.write(f"⏱️ **Time Taken:** {end_time - start_time:.2f} seconds")
```
